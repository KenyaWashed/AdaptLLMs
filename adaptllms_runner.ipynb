````python
import os, shutil 

src = "/kaggle/temp/huggingface/cache"
dst = "/kaggle/working/cache"

if os.path.exists(src):
    os.makedirs(dst, exist_ok=True)
    for item in os.listdir(src):
        s = os.path.join(src, item)
        d = os.path.join(dst, item)
        if os.path.isdir(s):
            shutil.copytree(s, d, dirs_exist_ok=True)
        else:
            shutil.copy2(s, d)
    print(f"Copied cache from {src} â†’ {dst}")
else:
    print(f"Source path not found: {src}")
````

# 4. API

````python
# === 1. Import ===
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch
import os
import random
from IPython.display import display, Markdown
import ipywidgets as widgets

# === 2. Thiáº¿t láº­p cache vÃ  load model ===
cache_candidates = ["/kaggle/working/cache", "/kaggle/temp/huggingface/cache"]
cache_dir = next((p for p in cache_candidates if os.path.exists(p)), "/kaggle/temp/huggingface/cache")
print(f"Using cache directory: {cache_dir}")

model_name = "facebook/opt-125m"

dtype = torch.float16 if torch.cuda.is_available() else torch.float32
device_map = "auto" if torch.cuda.is_available() else None

tokenizer = AutoTokenizer.from_pretrained(model_name, cache_dir=cache_dir)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    cache_dir=cache_dir,
    torch_dtype=dtype,
    device_map=device_map
)

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

# === 3. HÃ m sinh raw text ngáº«u nhiÃªn ===
def generate_random_raw_text():
    samples = [
        """Recent studies in biomedicine have shown a strong link between gut microbiota and brain function.
The gut-brain axis represents a bidirectional communication system involving neural, hormonal,
and immune pathways. Alterations in gut bacteria composition have been associated with
neurological disorders such as depression, anxiety, and Alzheimerâ€™s disease.
Understanding this relationship could lead to new therapeutic strategies for brain health.""",
        
        """Artificial intelligence is transforming medical imaging by automating the detection of diseases
from radiology scans. Deep neural networks can analyze thousands of X-ray or MRI images faster
than humans and can even highlight regions of interest for doctors to review.
Such AI-driven diagnostic systems hold great promise for early disease detection.""",
        
        """COVID-19 vaccines have demonstrated high efficacy rates in clinical trials across diverse populations.
However, vaccine hesitancy remains a significant barrier to achieving herd immunity.
Public health agencies are working to address misinformation and improve vaccine accessibility worldwide."""
    ]
    #return random.choice(samples)
    return "\n\n".join(samples)

# === 4. Giao diá»‡n nháº­p vÃ  xem raw text ===
raw_text_box = widgets.Textarea(
    value=generate_random_raw_text(),
    placeholder="Raw text hiá»ƒn thá»‹ táº¡i Ä‘Ã¢y...",
    description="RAW DATA:",
    layout=widgets.Layout(width="100%", height="220px")
)

generate_button = widgets.Button(
    description="Sinh raw text ngáº«u nhiÃªn",
    button_style="info",
    tooltip="Sinh Ä‘oáº¡n raw text ngáº«u nhiÃªn Ä‘á»ƒ test"
)

def on_generate_clicked(_):
    raw_text_box.value = generate_random_raw_text()

generate_button.on_click(on_generate_clicked)

# === 5. Ã” nháº­p cÃ¢u há»i ngÆ°á»i dÃ¹ng ===
question_box = widgets.Textarea(
    value="What is the main focus of this passage?",
    placeholder="Nháº­p cÃ¢u há»i hoáº·c yÃªu cáº§u cá»§a báº¡n á»Ÿ Ä‘Ã¢y...",
    description="User Question:",
    layout=widgets.Layout(width="100%", height="100px")
)

task_dropdown = widgets.Dropdown(
    options=[
        "Auto-detect",
        "Text Completion",
        "Summarization",
        "Question Answering",
        "Word-to-Text",
        "Natural Language Inference",
        "Paraphrase Detection",
        "Commonsense Reasoning",
    ],
    value="Auto-detect",
    description="Task:",
    style={"description_width": "initial"},
)

run_button = widgets.Button(
    description="ğŸš€ Run Inference",
    button_style="success",
    tooltip="Cháº¡y model vá»›i raw text vÃ  cÃ¢u há»i"
)

output_area = widgets.Output()

# === 6. HÃ m build prompt ===
def build_prompt(raw_text, question, task):
    if task == "Auto-detect":
        t = (raw_text + " " + question).lower()
        if "summary" in t:
            task = "Summarization"
        elif "premise" in t and "hypothesis" in t:
            task = "Natural Language Inference"
        elif "contradict" in t or "paraphrase" in t:
            task = "Paraphrase Detection"
        elif "reason" in t:
            task = "Commonsense Reasoning"
        elif "question" in t or "who" in t or "what" in t or "why" in t:
            task = "Question Answering"
        else:
            task = "Text Completion"

    # === chá»n prompt theo task ===
    if task == "Summarization":
        prompt = f"Summarize the following text:\n{raw_text}\n\nSummary:"
    elif task == "Question Answering":
        prompt = f"Answer the question based on the following text.\n\nText:\n{raw_text}\n\nQuestion: {question}\nAnswer:"
    elif task == "Word-to-Text":
        prompt = f"Generate a sentence using these biomedical terms:\n{raw_text}\n\nSentence:"
    elif task == "Natural Language Inference":
        prompt = f"Determine if the hypothesis is entailed by the premise.\n{raw_text}\n\nAnswer (Yes/No):"
    elif task == "Paraphrase Detection":
        prompt = f"Determine if the following two sentences mean the same:\n{raw_text}\n\nAnswer (Yes/No):"
    elif task == "Commonsense Reasoning":
        prompt = f"Use common sense reasoning to answer:\n{raw_text}\n\nQuestion: {question}\nAnswer:"
    else:
        prompt = f"{raw_text.strip()}\n\nQuestion: {question}\nAnswer:"

    return prompt, task

# === 7. HÃ m cháº¡y inference ===
# def run_inference(_):
#     with output_area:
#         output_area.clear_output()
#         raw_text = raw_text_box.value.strip()
#         question = question_box.value.strip()
#         task = task_dropdown.value

#         if not raw_text:
#             print("Vui lÃ²ng nháº­p hoáº·c sinh raw text.")
#             return
#         if not question:
#             print("Vui lÃ²ng nháº­p cÃ¢u há»i.")
#             return

#         prompt, chosen_task = build_prompt(raw_text, question, task)

#         display(Markdown(f"### Task: {chosen_task}"))
#         display(Markdown(f"**Prompt:**\n```\n{prompt}\n```"))
#         display(Markdown("### Model Output:"))

#         result = generator(prompt, max_new_tokens=200, do_sample=True, top_p=0.9, temperature=0.7)
#         print(result[0]["generated_text"].replace(prompt, "").strip())

def run_inference(_):
    with output_area:
        output_area.clear_output()
        raw_text = raw_text_box.value.strip()
        question = question_box.value.strip()
        task = task_dropdown.value

        if not raw_text:
            print("Vui lÃ²ng nháº­p hoáº·c sinh raw text.")
            return
        if not question:
            print("Vui lÃ²ng nháº­p cÃ¢u há»i.")
            return

        prompt, chosen_task = build_prompt(raw_text, question, task)

        display(Markdown(f"### Task: {chosen_task}"))
        display(Markdown(f"**Prompt:**\n```\n{prompt}\n```"))
        display(Markdown("### Model Output:"))

        # === Cháº¡y inference ===
        result = generator(
            prompt,
            max_new_tokens=200,
            do_sample=True,
            top_p=0.9,
            temperature=0.7,
            repetition_penalty=1.2,
            no_repeat_ngram_size=3,
        )

        output_text = result[0]["generated_text"].replace(prompt, "").strip()

        # === Lá»c chá»‰ láº¥y 1 láº§n "Answer" Ä‘áº§u tiÃªn ===
        if "Answer:" in output_text:
            output_text = output_text.split("Answer:")[-1].strip()

        # === Cáº¯t bá» náº¿u cÃ³ nhiá»u dÃ²ng trÃ¹ng láº·p ===
        lines = output_text.split("\n")
        cleaned = []
        seen = set()
        for line in lines:
            line_clean = line.strip()
            if line_clean and line_clean not in seen:
                cleaned.append(line_clean)
                seen.add(line_clean)
        final_answer = " ".join(cleaned)

        print(final_answer)


run_button.on_click(run_inference)

# === 8. Hiá»ƒn thá»‹ giao diá»‡n Ä‘áº§y Ä‘á»§ ===
display(
    Markdown("## RAW TEXT PREVIEW"),
    raw_text_box,
    generate_button,
    Markdown("## USER QUESTION"),
    question_box,
    task_dropdown,
    run_button,
    output_area
)
````
